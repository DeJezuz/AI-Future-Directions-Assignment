```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error
import numpy as np # Added this import

np.random.seed(7)
n_plots = 200
days = 120
rows = []
for plot in range(n_plots):
    soil_type = np.random.choice([0,1,2])
    cultivar = np.random.choice([0,1,2])
    temp = 20 + 5*np.sin(np.linspace(0,2*np.pi,days)) + np.random.normal(0,1.5,days)
    moisture = 30 + 15*np.random.beta(2,5,days)
    ndvi = 0.3 + 0.5*np.clip(np.linspace(0,1,days)+np.random.normal(0,0.05,days),0,1)
    feats = {
        'soil_type':soil_type,
        'cultivar':cultivar,
        'temp':temp,
        'moisture':moisture,
        'ndvi':ndvi
    }
    rows.append(feats) # Appended the generated features to rows
```
